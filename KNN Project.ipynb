```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('KNN_Project_Data.csv')
```

```python
df.head()
```

```python
sns.pairplot(df, hue='TARGET CLASS', palette='coolwarm', diag_kws={'bw': 0.2})
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# fit scaler into features

scaler.fit(df.drop('TARGET CLASS', axis = 1))
```

```python
# transform the features to scaled version

scaled_features = scaler.transform(df.drop('TARGET CLASS', axis = 1))
```

```python
# convert scaled features to df

df_feat = pd.DataFrame(scaled_features, columns = df.columns[:-1])
```

```python
df_feat.head(4)
```

```python
# TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split
X = df_feat 
y = df['TARGET CLASS']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Import KNN from scikit learn
# create a knn model with n_neighbors = 1
# fit the model to the training data

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

```python
#Predictions and evaluations

pred = knn.predict(X_test)
```

```python
pred
```

```python
#create a confusion matrix and clas report


from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
# elbow method to choose a better k value

error_rate = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50), error_rate, color = 'blue', linestyle ='dashed',
         marker ='o', markerfacecolor ='red', markersize=8)
plt.title('Error Rate vs K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors = 40)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
